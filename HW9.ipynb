# Problem 1: Probability

```python
import random

def probability():
    number_games = int(input('How many times would you like to play? '))
    wins_changed_guess = 0
    print()
    print()
    print('The Door With The Prize     ', 'The Door You Guessed   ', 'The Door That Was Removed   ', 'Your New Guess     ')
    
    for current_game in range(number_games):
        door_prize = random.randint(1, 3)
        door_your_guess = random.randint(1, 3)
        door_removed = random.randint(1, 3)
        while door_removed == door_prize or door_removed == door_your_guess:
            door_removed = random.randint(1, 3)

        change_guess = random.randint(1, 3)
        while change_guess == door_removed or change_guess == door_your_guess:
            change_guess = random.randint(1, 3)

        if change_guess == door_prize:
            wins_changed_guess += 1

        print(door_prize, '                          ', door_your_guess, '                     ', door_removed, '                          ', change_guess)

    win_change = (wins_changed_guess)/(number_games)
    print()
    print()
    print('The probability of winning if you change your guess = %.2f' %round(win_change, 2))
    print('The probability of winning if you do not change your guess = %.2f' %round(1 - win_change, 2))
    
probability()
```

# Problem 2: Radioactive Decay

```python
from random import random
from numpy import arange
from pylab import plot,xlabel,ylabel,show

# Constants
NTl = 100             # Initial Number of thallium atoms
NPb = 0               # Number of lead atoms
tau = 3.053*60        # Half life of thallium in seconds
h = 1.0               # Size of time-step in seconds
p = 1 - 2**(-h/tau)   # Probability of decay in one step
tmax = 1000           # Total time

# Lists of plot points
tpoints = arange(0.0,tmax,h) # make time array
Tlpoints = [] # empty array to strore # of Tl atoms at each time step
Pbpoints = [] # empty array to strore # of Pb atoms at each time step

# Main loop
for t in tpoints:    # for time 0-1000 sec
    Tlpoints.append(NTl) # append starting # of Tl atoms
    Pbpoints.append(NPb) # append starting # of Pb atoms

    # Calculate the number of atoms that decay
    decay = 0
    for i in range(NTl): # determine number of atoms that decay
        if random() < p:
            decay += 1
    NTl -= decay # subtract number of decayed Parent atoms 
    NPb += decay # add number of decayed Daughter atoms 

# Make the graph
plot(tpoints,Tlpoints, c='m') # plot Tl vs. time
plot(tpoints,Pbpoints, c='b') # plot Pb vs. time
xlabel("Time")
ylabel("Number of atoms")
show()
```
